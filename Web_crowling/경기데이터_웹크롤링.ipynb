```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromiumService
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.core.os_manager import ChromeType
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC

# ChromeDriver 설정 및 WebDriver 초기화
def initialize_driver():
    driver = webdriver.Chrome(
        service=ChromiumService(ChromeDriverManager(chrome_type=ChromeType.GOOGLE).install())
    )
    driver.get('https://data.kleague.com/')
    return driver

# 프레임 전환 함수
def switch_to_frame(driver):
    """프레임 전환을 시도합니다."""
    try:
        frames = driver.find_elements(By.TAG_NAME, "frame")
        print(f"총 {len(frames)}개의 프레임이 있습니다.")
        if len(frames) > 1:
            driver.switch_to.frame(frames[1])  # 두 번째 프레임으로 전환
            print("두 번째 프레임으로 전환 완료")
        else:
            print("두 번째 프레임을 찾을 수 없습니다.")
    except Exception as e:
        print(f"프레임 전환 중 오류 발생: {e}")

# Match Center 요소 클릭
def click_match_center(driver):
    """'Match Center' 요소를 클릭합니다."""
    try:
        match_center_element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, 'mcCenter'))
        )
        match_center_element.click()
        time.sleep(3)
        print("'Match Center' 클릭 성공")
    except Exception as e:
        print(f"'Match Center' 클릭 중 오류 발생: {e}")

# PLAYER STATS 탭 클릭
def click_player_stats(driver):
    """'PLAYER STATS' 탭을 클릭합니다."""
    try:
        player_stats_tab = driver.find_element(By.XPATH, "//*[@id='contentsLayer']/div/div/ul/li[3]")
        player_stats_tab.click()
        time.sleep(3)
        print("'PLAYER STATS' 탭 클릭 성공")
    except Exception as e:
        print(f"'PLAYER STATS' 탭 클릭 중 오류 발생: {e}")

# 연도 선택 함수
def select_year(driver, year_value):
    year_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selMeetYear']"))
    year_dropdown.select_by_value(year_value)
    print(f"연도 선택: {year_value}")
    time.sleep(2)

# 리그 선택 함수
def select_league(driver):
    """리그를 선택합니다 (value='1')."""
    try:
        league_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selMeetSeq']"))
        league_dropdown.select_by_value("1")  # value가 1인 리그 선택
        print("리그 선택: value=1")
        time.sleep(2)
    except Exception as e:
        print(f"리그 선택 중 오류 발생: {e}")

# 날짜 선택 함수
def select_date(driver, date_value):
    date_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selRoundId']"))
    date_dropdown.select_by_value(str(date_value))
    print(f"날짜 선택: {date_value}")
    time.sleep(2)

# 시합 선택 함수
def select_match(driver, match_value):
    """시합 선택 함수. match_value에 따라 시합을 선택합니다."""
    try:
        match_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selGameId']"))
        match_dropdown.select_by_value(match_value)
        print(f"시합 선택: {match_value}")
        time.sleep(2)
    except Exception as e:
        print(f"시합 선택 중 오류 발생: {e}")

# 경기 정보 추출 함수 (홈팀, 어웨이팀, 세부정보 포함)
def extract_match_info(driver):
    """홈팀, 어웨이팀 정보 및 경기 세부정보를 하나의 데이터프레임으로 반환합니다."""
    try:
        # 홈팀 및 어웨이팀 정보 추출
        home_team = driver.find_element(By.XPATH, "//*[@id='frm']/div[2]/div[2]/span[1]").text
        away_team = driver.find_element(By.XPATH, "//*[@id='frm']/div[3]/div[1]/span[3]").text

        # 경기 세부정보 추출
        features = ["Home Team", "Away Team"]
        values = [home_team, away_team]

        xpaths = [
            ("//*[@id='frm']/div[1]/div/ul/li[1]/span[1]", "//*[@id='frm']/div[1]/div/ul/li[1]/span[2]"),
            ("//*[@id='frm']/div[1]/div/ul/li[2]/span[1]", "//*[@id='frm']/div[1]/div/ul/li[2]/span[2]"),
            ("//*[@id='frm']/div[1]/div/ul/li[3]/span[1]", "//*[@id='frm']/div[1]/div/ul/li[3]/span[2]"),
            ("//*[@id='frm']/div[1]/div/ul/li[4]/span[1]", "//*[@id='frm']/div[1]/div/ul/li[4]/span[2]"),
            ("//*[@id='frm']/div[1]/div/ul/li[5]/span[1]", "//*[@id='frm']/div[1]/div/ul/li[5]/span[2]")
        ]

        for feature_xpath, value_xpath in xpaths:
            feature = driver.find_element(By.XPATH, feature_xpath).text
            value = driver.find_element(By.XPATH, value_xpath).text
            features.append(feature)
            values.append(value)

# 데이터프레임 생성
        df = pd.DataFrame([values], columns=features)

#         # CSV 파일로 저장
#         save_path = "/Users/chjun/Desktop/데이터마이닝/detail_info_data"
#         filename = os.path.join(save_path, f"{year_value}_{match_text}_날씨.csv")
#         df.to_csv(filename, index=False, encoding='utf-8-sig')
#         print(f"CSV 파일 저장 완료: {filename}")

        return df

    except Exception as e:
        print(f"경기 정보 추출 중 오류 발생: {e}")
        return pd.DataFrame()

# 전체 데이터 추출 및 순회 함수
def extract_all_data():
    driver = initialize_driver()

    try:
        switch_to_frame(driver)  # 프레임 전환
        click_match_center(driver)  # 'Match Center' 클릭
        click_player_stats(driver)  # 'PLAYER STATS' 탭 클릭

        all_data = pd.DataFrame()

        # 연도 순회: 2021, 2022, 2023
        for year_value in ["2021", "2022", "2023"]:
            select_year(driver, year_value)
            select_league(driver)  # 리그 선택
            
            # 날짜 순회: 1부터 38까지
            for date_value in range(1, 39):
                select_date(driver, date_value)
                
                # 시합 순회: 첫 6개의 시합 선택
                match_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selGameId']"))
                match_values = [option.get_attribute("value") for option in match_dropdown.options if option.get_attribute("value")]

                for match_value in match_values:
                    select_match(driver, match_value)
                    df = extract_match_info(driver)
                    print(df)  # 시합별로 데이터프레임 출력
                    all_data = pd.concat([all_data, df], ignore_index=True)

        # 최종 데이터프레임 출력
        print(all_data)

    except Exception as e:
        print(f"데이터 추출 중 오류 발생: {e}")
    finally:
        driver.quit()

# 데이터 추출 실행
extract_all_data()
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromiumService
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.core.os_manager import ChromeType
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC

# CSV 파일 경로 설정
file_path = "/Users/chjun/Desktop/데이터마이닝/detail_info_data/match_detail_data_info.csv"
base_path = "/Users/chjun/Desktop/데이터마이닝/detail_info_data"

# ChromeDriver 설정 및 WebDriver 초기화
def initialize_driver():
    driver = webdriver.Chrome(
        service=ChromiumService(ChromeDriverManager(chrome_type=ChromeType.GOOGLE).install())
    )
    driver.get('https://data.kleague.com/')
    return driver

# 프레임 전환 함수
def switch_to_frame(driver):
    """프레임 전환을 시도합니다."""
    try:
        frames = driver.find_elements(By.TAG_NAME, "frame")
        print(f"총 {len(frames)}개의 프레임이 있습니다.")
        if len(frames) > 1:
            driver.switch_to.frame(frames[1])  # 두 번째 프레임으로 전환
            print("두 번째 프레임으로 전환 완료")
        else:
            print("두 번째 프레임을 찾을 수 없습니다.")
    except Exception as e:
        print(f"프레임 전환 중 오류 발생: {e}")

# Match Center 요소 클릭
def click_match_center(driver):
    """'Match Center' 요소를 클릭합니다."""
    try:
        match_center_element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, 'mcCenter'))
        )
        match_center_element.click()
        time.sleep(3)
        print("'Match Center' 클릭 성공")
    except Exception as e:
        print(f"'Match Center' 클릭 중 오류 발생: {e}")

# PLAYER STATS 탭 클릭
def click_player_stats(driver):
    """'PLAYER STATS' 탭을 클릭합니다."""
    try:
        player_stats_tab = driver.find_element(By.XPATH, "//*[@id='contentsLayer']/div/div/ul/li[3]")
        player_stats_tab.click()
        time.sleep(3)
        print("'PLAYER STATS' 탭 클릭 성공")
    except Exception as e:
        print(f"'PLAYER STATS' 탭 클릭 중 오류 발생: {e}")

# 연도 선택 함수
def select_year(driver, year_value):
    year_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selMeetYear']"))
    year_dropdown.select_by_value(year_value)
    print(f"연도 선택: {year_value}")
    time.sleep(2)

# 리그 선택 함수
def select_league(driver):
    """리그를 선택합니다 (value='1')."""
    try:
        league_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selMeetSeq']"))
        league_dropdown.select_by_value("1")  # value가 1인 리그 선택
        print("리그 선택: value=1")
        time.sleep(2)
    except Exception as e:
        print(f"리그 선택 중 오류 발생: {e}")

# 날짜 선택 함수
def select_date(driver, date_value):
    date_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selRoundId']"))
    date_dropdown.select_by_value(str(date_value))
    print(f"날짜 선택: {date_value}")
    time.sleep(2)

# 시합 선택 함수
def select_match(driver, match_value):
    """시합 선택 함수. match_value에 따라 시합을 선택합니다."""
    try:
        match_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selGameId']"))
        match_dropdown.select_by_value(match_value)
        print(f"시합 선택: {match_value}")
        time.sleep(2)
    except Exception as e:
        print(f"시합 선택 중 오류 발생: {e}")

# 점수 추가 함수
def add_scores_to_data(driver, data):
    """홈팀 점수와 어웨이팀 점수를 추가합니다."""
    try:
        # 프레임 전환
        switch_to_frame(driver)
        click_match_center(driver)
        click_player_stats(driver)

        # 점수 추가 작업
        print("홈팀 점수 및 어웨이팀 점수 열 추가 중...")
        home_scores = []
        away_scores = []

        # 연도 순회: 2021, 2022, 2023
        for year_value in ["2021", "2022", "2023"]:
            select_year(driver, year_value)
            select_league(driver)  # 리그 선택

            # 날짜 순회: 1부터 38까지
            for date_value in range(1, 39):
                select_date(driver, date_value)

                # 시합 순회: 모든 시합 선택
                match_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selGameId']"))
                match_values = [option.get_attribute("value") for option in match_dropdown.options if option.get_attribute("value")]

                for match_value in match_values:
                    select_match(driver, match_value)
                    try:
                        # 홈팀 점수 및 어웨이팀 점수 추출
                        home_score = driver.find_element(By.XPATH, "//*[@id='frm']/div[2]/div[2]/span[2]").text
                        away_score = driver.find_element(By.XPATH, "//*[@id='frm']/div[3]/div[1]/span[2]").text
                        home_scores.append(home_score)
                        away_scores.append(away_score)
                        print(f"경기: 홈팀 점수 - {home_score}, 어웨이팀 점수 - {away_score}")
                    except Exception as e:
                        print(f"점수 조회 중 오류 발생: {e}")
                        home_scores.append(None)
                        away_scores.append(None)

            # 연도별 데이터프레임에 점수 열 추가 및 중간 저장
            data['Home Team Score'] = home_scores
            data['Away Team Score'] = away_scores
            save_path = os.path.join(base_path, f"{year_value}_match_detail_data_info.csv")
            data.to_csv(save_path, index=False, encoding='utf-8-sig')
            print(f"{year_value}년의 데이터가 CSV 파일로 중간 저장되었습니다: {save_path}")

    except Exception as e:
        print(f"점수 추가 중 오류 발생: {e}")

# CSV 파일 불러오기 및 점수 추가
def update_csv_with_scores():
    print("CSV 파일 불러오는 중...")
    try:
        data = pd.read_csv(file_path)
        print("CSV 파일이 성공적으로 불러와졌습니다.")
    except Exception as e:
        print(f"CSV 파일 불러오기 중 오류 발생: {e}")
        return

    # 드라이버 초기화 및 점수 추가
    driver = initialize_driver()
    try:
        add_scores_to_data(driver, data)
    finally:
        driver.quit()

    # 수정된 데이터 출력
    print("수정된 데이터:")
    print(data)

    # 최종적으로 전체 데이터를 한 번 더 CSV로 저장
    print("CSV 파일 최종 업데이트 중...")
    try:
        data.to_csv(file_path, index=False, encoding='utf-8-sig')
        print(f"CSV 파일이 성공적으로 최종 업데이트되었습니다: {file_path}")
    except Exception as e:
        print(f"CSV 파일 저장 중 오류 발생: {e}")

# 점수 추가 및 CSV 업데이트 실행
update_csv_with_scores()
```

```python
import time
import pandas as pd
import os
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromiumService
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.core.os_manager import ChromeType
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC

# CSV 파일 경로 설정
file_path = "/Users/chjun/Desktop/데이터마이닝/detail_info_data/match_detail_data_info.csv"

# ChromeDriver 설정 및 WebDriver 초기화
def initialize_driver():
    driver = webdriver.Chrome(
        service=ChromiumService(ChromeDriverManager(chrome_type=ChromeType.GOOGLE).install())
    )
    driver.get('https://data.kleague.com/')
    return driver

# 프레임 전환 함수
def switch_to_frame(driver):
    """프레임 전환을 시도합니다."""
    try:
        frames = driver.find_elements(By.TAG_NAME, "frame")
        if len(frames) > 1:
            driver.switch_to.frame(frames[1])
    except Exception as e:
        print(f"프레임 전환 중 오류 발생: {e}")

# Match Center 요소 클릭
def click_match_center(driver):
    try:
        match_center_element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.ID, 'mcCenter'))
        )
        match_center_element.click()
        time.sleep(3)
    except Exception as e:
        print(f"'Match Center' 클릭 중 오류 발생: {e}")

# PLAYER STATS 탭 클릭
def click_player_stats(driver):
    try:
        player_stats_tab = driver.find_element(By.XPATH, "//*[@id='contentsLayer']/div/div/ul/li[3]")
        player_stats_tab.click()
        time.sleep(3)
    except Exception as e:
        print(f"'PLAYER STATS' 탭 클릭 중 오류 발생: {e}")

# 연도 선택 함수
def select_year(driver, year_value):
    year_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selMeetYear']"))
    year_dropdown.select_by_value(year_value)
    time.sleep(2)

# 리그 선택 함수
def select_league(driver):
    try:
        league_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selMeetSeq']"))
        league_dropdown.select_by_value("1")
        time.sleep(2)
    except Exception as e:
        print(f"리그 선택 중 오류 발생: {e}")

# 날짜 선택 함수
def select_date(driver, date_value):
    date_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selRoundId']"))
    date_dropdown.select_by_value(str(date_value))
    time.sleep(2)

# 시합 선택 함수
def select_match(driver, match_value):
    try:
        match_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selGameId']"))
        match_dropdown.select_by_value(match_value)
        time.sleep(2)
    except Exception as e:
        print(f"시합 선택 중 오류 발생: {e}")

# 점수 추가 함수
def add_scores_to_data(driver):
    """홈팀 점수와 어웨이팀 점수를 차례로 추가하고 저장합니다."""
    # CSV 파일 불러오기
    data = pd.read_csv(file_path)
    
    # 프레임 전환 및 기본 클릭 작업
    switch_to_frame(driver)
    click_match_center(driver)
    click_player_stats(driver)

    # 연도 순회: 2021, 2022, 2023
    for year_value in ["2021", "2022", "2023"]:
        select_year(driver, year_value)
        select_league(driver)  # 리그 선택

        # 날짜 순회: 1부터 38까지
        for date_value in range(1, 39):
            select_date(driver, date_value)

            # 시합 순회: 모든 시합 선택
            match_dropdown = Select(driver.find_element(By.XPATH, "//*[@id='selGameId']"))
            match_values = [option.get_attribute("value") for option in match_dropdown.options if option.get_attribute("value")]

            for match_value in match_values:
                select_match(driver, match_value)
                try:
                    # 홈팀 점수 및 어웨이팀 점수 추출
                    home_score = driver.find_element(By.XPATH, "//*[@id='frm']/div[2]/div[2]/span[2]").text
                    away_score = driver.find_element(By.XPATH, "//*[@id='frm']/div[3]/div[1]/span[2]").text
                    print(f"경기: 홈팀 점수 - {home_score}, 어웨이팀 점수 - {away_score}")

                    # 새 점수를 데이터프레임에 추가
                    data = data.append({
                        "Home Team Score": home_score,
                        "Away Team Score": away_score
                    }, ignore_index=True)

                    # CSV 파일에 실시간 저장
                    data.to_csv(file_path, index=False, encoding='utf-8-sig')
                    print("CSV 파일이 성공적으로 업데이트되었습니다.")
                except Exception as e:
                    print(f"점수 조회 중 오류 발생: {e}")

# 실행 함수
def update_csv_with_scores():
    driver = initialize_driver()
    try:
        add_scores_to_data(driver)
    finally:
        driver.quit()

# 실행
update_csv_with_scores()
```
